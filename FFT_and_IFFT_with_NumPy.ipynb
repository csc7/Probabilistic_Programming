```python
# External dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define text size for plots
import matplotlib
matplotlib.rc('xtick', labelsize=16) 
matplotlib.rc('ytick', labelsize=16)
matplotlib.rcParams.update({'axes.labelsize': 16})
```

```python
# Signal

f = 20  # Frequency
x = np.linspace(0, 100, 100000) 
f_t = 10 * np.sin(2 * np.pi * f * x)

# Plot
plt.plot(f_t[0:500]) # First 500 samples for better visualization
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# FFT

N = len(f_t)
sample_spacing = (x[1] - x[0])

FFT_f_t = np.fft.fft(f_t)
FFT_f_t_freqs = np.fft.fftfreq(N, sample_spacing)

# Plot
plt.plot(FFT_f_t_freqs, FFT_f_t)
plt.xlim(-100, 100)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
# IFFT

IFFT_f_t = np.fft.ifft(FFT_f_t).real

# Plot
plt.plot(IFFT_f_t[0:500])
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show
```
